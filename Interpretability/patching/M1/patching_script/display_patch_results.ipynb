### Visualize stats:

Three types of overlap are calculated:

1. with respect to predicted lengths,

2. w.r.t. to real length,

3. intersection over union (attempts at joining the two previous ones into one score).

Ideal outcome would be if both 1., 2. overlaps were high (=> intervals are almost identital).

If both are 0, there is no overlap.

If w.r.t. pred is high and w.r.t. real is low, it means that the chosen patch size was too small.

```python
def print_overlap_stats(df, column, print_str):
    print(f'Stats {print_str}:')
    pred_mean = df[column].mean()
    print(f'Mean: {pred_mean}')
    pred_med = df[column].median()
    print(f'Med: {pred_med}')
    pred_max = df[column].max()
    print(f'Max: {pred_max}')
    pred_min = df[column].min()
    print(f'Min: {pred_min}')
    print()
    
print_overlap_stats(df, 'overlap_pred', 'w.r.t. predicted length (how much of the patch is inside knot core)')
print_overlap_stats(df, 'overlap_real', 'w.r.t. real length (how much of the knot core was discovered)')
print_overlap_stats(df, 'overlap', 'intersection over union (both metrics together)')
print_overlap_stats(df, 'min_pred', 'average prediction score drop')
df['drop_difference'] = df.apply(lambda row: row['sequence_pred'] - row['min_pred'], axis = 1)
print_overlap_stats(df, 'drop_difference', 'average drop difference')
```

Get percentage of overlaps over some specified threshold:

```python
PRED_THRESHOLD = 0.6
REAL_THRESHOLD = 0.5
BOTH_THRESHOLD = 0.2

def get_simple_results_positive(df):
    simple_results = []
    sequences_count = df.shape[0]
    reasonable_wrt_pred_count = df.loc[df['overlap_pred'] >= PRED_THRESHOLD].shape[0]
    reasonable_wrt_real_count = df.loc[df['overlap_real'] >= REAL_THRESHOLD].shape[0]
    reasonable_wrt_count = df.loc[df['overlap'] > PRED_THRESHOLD].shape[0]

    simple_results.append({'patch_size': PATCH_SIZE,
                           'count w.r.t. pred': reasonable_wrt_pred_count,
                           'percentage w.r.t. pred': reasonable_wrt_pred_count/sequences_count,
                           'count w.r.t. real': reasonable_wrt_real_count,
                           'percentage w.r.t. real': reasonable_wrt_real_count/sequences_count,
                           'count w.r.t. intersection': reasonable_wrt_count,
                           'percentage w.r.t. intersection': reasonable_wrt_count/sequences_count})

    return pd.DataFrame(simple_results)

get_simple_results_positive(df)
```

```python
import matplotlib.pyplot as plt

def plot_overlap_scores(df):
    plt.rcParams['figure.figsize'] = (200, 20)
    ax = df.plot(x='index', y=['min_pred', 'overlap_real'], kind='bar', ylim=(-0.02,1.02)) # ylim to force y-axis scale to be the same as it is for the other graphs
    ax.legend(['pred', 'real'])
    ax.set_xlabel(f'Seq index')
    ax.set_ylabel('Percentage of overlap')
    plt.show()
        
    ax = df.plot(x='index', y=['overlap'], kind='bar', ylim=(-0.02,1.02))
    ax.set_xlabel(f'Seq index')
    ax.set_ylabel('Intersection over union')
    plt.show()

df['index'] = df.index
plot_overlap_scores(df)
```

Display knot core vs. patch overlap (as coloured string):

```python
from termcolor import colored

# https://stackoverflow.com/questions/16816013/is-it-possible-to-print-using-different-colors-in-ipythons-notebook
def format_seq_str(seq_str, knot_start, knot_end, patch_start, patch_end):
    # there are multiple cases, go over each:
    
    # a) no overlap, patch on left:    SSPPSKKKKSSSSS
    if (patch_start <= knot_start) & (patch_end <= knot_start):
        return seq_str[:patch_start] + colored(seq_str[patch_start:patch_end], 'yellow') + seq_str[patch_end:knot_start] + colored(seq_str[knot_start:knot_end], 'green') + seq_str[knot_end:]
    # b) overlap from left:            SSPPSKKKKSSSSS
    elif (patch_start <= knot_start) & (patch_end <= knot_end):
        return seq_str[:patch_start] + colored(seq_str[patch_start:knot_start], 'yellow') + colored(seq_str[knot_start:patch_end], 'blue') + colored(seq_str[patch_end:knot_end], 'green') + seq_str[knot_end:]
    # c) entire overlap:               SSSSSKPPKSSSSS
    elif (knot_start <= patch_start) & (knot_end >= patch_end):
        return seq_str[:knot_start] + colored(seq_str[knot_start:patch_start], 'green') + colored(seq_str[patch_start:patch_end], 'blue') + colored(seq_str[patch_end:knot_end], 'green') + seq_str[knot_end:]
    # c2) entire overlap, bigger patch: SSSSSKPPKSSSSS
    elif (patch_start <= knot_start) & (patch_end >= knot_end):
        return seq_str[:patch_start] + colored(seq_str[patch_start:knot_start], 'yellow') + colored(seq_str[knot_start:knot_end], 'blue') + colored(seq_str[knot_end:patch_end], 'yellow') + seq_str[patch_end:]
    # d) no overlap, patch on right:   SSSSSKKKKSSPPS
    elif (patch_start >= knot_end):
        return seq_str[:knot_start] + colored(seq_str[knot_start:knot_end], 'green') + seq_str[knot_end:patch_start] + colored(seq_str[patch_start:patch_end], 'yellow') + seq_str[patch_end:]
    # e) overlap from right:           SSSSSKKKPPSSSS
    elif (knot_start <= patch_start) & (knot_end <= patch_end):
        return seq_str[:knot_start] + colored(seq_str[knot_start:patch_start], 'green') + colored(seq_str[patch_start:knot_end], 'blue') + colored(seq_str[knot_end:patch_end], 'yellow') + seq_str[patch_end:]
    else:
        return 'Error in boundaries'
    
print(f'Knot core is highligted in {colored("green", "green")} and the patch is {colored("blue", "blue")} (if they do not overlap, the patch is {colored("yellow", "yellow")}).')
```

Show all drops (minimum is highlighted in red):

```python
DECISION_THRESHOLD = 0.5


def plot_drops(preds_list, min_index, min_value):
    # https://stackoverflow.com/questions/74013277/count-number-of-peaks-when-using-pyplot
    positions = pd.DataFrame([_ for _ in range(len(preds_list))], columns=['position'])
    positions['prediction'] = preds_list
    peaks = positions[positions['prediction'] < DECISION_THRESHOLD]
    plt.rcParams['figure.figsize'] = (5, 3)
    plt.plot(positions['position'], positions['prediction'], label='Patched prediction')
    plt.plot(positions, [DECISION_THRESHOLD] * len(positions), c='k')
    plt.scatter(peaks.index, peaks['prediction'], c='y', marker='X', label='Prediction < DECISION_THRESHOLD') # marker: red cross
    plt.scatter(min_index+1, min_value, c='r', marker='X', label='Minimum prediction') # marker: red cross
    plt.ylim([-0.05, 1.02])
    plt.xlabel('Patch position in the protein sequence')
    plt.ylabel('Prediction value')
    plt.legend(loc='best')
    plt.show()
    return len(peaks)


def plot_core_min_overlap(preds_list, min_index, min_value, knot_start, knot_end, seq_len):
    fig, ax = plt.subplots()
    ax.set_yticks([0.0, 0.5, 1.0])
    
    # plot knot core:
    # knot_scores = [0.0 if _ >= knot_start and _ <= (knot_end - PATCH_SIZE) else 1.0 for _ in range(seq_len - PATCH_SIZE)]
    knot_scores = [0.0 if _ >= (knot_start - PATCH_SIZE) and _ <= knot_end else 1.0 for _ in range(seq_len - PATCH_SIZE)]
    plt.plot([_ for _ in range(seq_len - PATCH_SIZE)], knot_scores, color='#67cf67', label='Knot core')
    
    # https://stackoverflow.com/questions/74013277/count-number-of-peaks-when-using-pyplot
    positions = pd.DataFrame([_ for _ in range(len(preds_list))], columns=['position'])
    positions['prediction'] = preds_list
    #peaks = positions[positions['prediction'] < THRESHOLD]
    plt.plot(positions['position'], positions['prediction'], label='Patched prediction')
    plt.plot(positions['position'], [DECISION_THRESHOLD] * len(positions), c='k', label='Prediction threshold')
    
    # plot concrete patch crosses:
    plt.scatter([0], positions['prediction'][2], c='y', marker='o', label='First patch')
    plt.scatter(min_index+1, min_value, c='r', marker='o', label='Minimum patch')
    
    plt.ylim([-0.05, 1.02])
    plt.xlabel('Patch position in the protein sequence')
    plt.ylabel('Prediction value')
    plt.legend(loc='best')
    plt.show()


for i in range(8):
    print(f'i={i}')
    print(f'Seq: {format_seq_str(df.loc[i]["sequence_str"], df.loc[i]["knot_start"], df.loc[i]["knot_end"], df.loc[i]["min_start"], df.loc[i]["min_end"])} ({len(df.loc[i]["sequence_str"])})')
    print(f'Raw pred: {df.loc[i]["sequence_pred"]} knot_core=({df.loc[i]["knot_start"]}, {df.loc[i]["knot_end"]}), Min pred: {df.loc[i]["min_pred"]} min_patch=({df.loc[i]["min_start"]}, {df.loc[i]["min_end"]})')
    print(f'Overlap: {df.loc[i]["overlap_pred"]}')
    plot_drops(df.loc[i]['patched_preds'], df.loc[i]['min_start'], df.loc[i]['min_pred'])
    plot_core_min_overlap(df.loc[i]['patched_preds'], df.loc[i]['min_start'], df.loc[i]['min_pred'], df.loc[i]['knot_start'], df.loc[i]['knot_end'], len(df.loc[i]['sequence_str']))
```
