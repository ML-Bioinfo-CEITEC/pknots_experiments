## Transform CSV format of the new data

- *for Jan 2023 data version*

Goal: Take 1000 random samples from each family (`FamilyName` column) + reformat the knot core interval info.

```python
HF_DATASET = 'EvaKlimentova/knots_AF'
OUTPUT = '/home/jovyan/data/proteins/raw_new_data_with_core_intervals.csv'
```

```python
from datasets import load_dataset

dataset = load_dataset(HF_DATASET)
dataset
```

```python
import pandas as pd

df_test = dataset['test'].to_pandas()
df_test
```

```python
df_test = df_test[['ID', 'uniprotSequence', 'globalMetricValue', 'Domain_architecture', 'InterPro', 'Max_Topology', 'Length', 'Knot Core', 'label', 'FamilyName']].copy()
df_test
```

## Clean & reformat the data:

```python
df_test['Length'] = pd.to_numeric(df_test['Length'])
df_test['Length'] = df_test['Length'].apply(lambda _: int(_))
df_test['FamilyName'] = df_test['FamilyName'].astype('str')
```

```python
# drop all sequences with sequence_length > 500:
df_test.drop(df_test.loc[df_test['Length'] > 500].index, inplace=True)
df_test
```

```python
df_test.dropna(subset=['Knot Core'], inplace=True)  # remove NaN values (+ only knotted sequences have something in the 'Knot Core' column)
df_test
```

Reformat the knot core interval:

```python
df_test['knot_start'] = df_test['Knot Core'].apply(lambda _: int(_.replace('(', '').replace(')', '').replace(',', '').split()[0])-1)  # -1 because knot core info is indexed from 1 and not 0 (patch position is indexed from 0)
df_test['knot_end'] = df_test['Knot Core'].apply(lambda _: int(_.replace('(', '').replace(')', '').replace(',', '').split()[1])-1)
df_test.drop('Knot Core', axis=1, inplace=True)
df_test
```

```python
df_test['knot_len'] = df_test['knot_end'] - df_test['knot_start']
df_test
```

```python
df_test['core_percentage'] = (df_test['knot_len'] * 100) / df_test['Length']
df_test
```

### Sample the data:

Take 1000 samples from each family *(or less if the family contains fewer samples)*.

```python
families = list(df_test['FamilyName'].unique())
families
```

```python
df_test[df_test['FamilyName'] == 'SPOUT'].sample(1000, random_state=42).reset_index(drop=True)
```

```python
dfs = []

for family_name in families:
    family_size = len(df_test[df_test['FamilyName'] == family_name])
    family_sample_size = 1000 if family_size >= 1000 else family_size
    df_tmp = df_test[df_test['FamilyName'] == family_name].sample(family_sample_size, random_state=42).reset_index(drop=True)
    dfs.append(df_tmp)
    
df_test_reduced = pd.concat(dfs)
df_test_reduced = df_test_reduced.rename(columns={'ID': 'seq_id', 
                                                  'uniprotSequence': 'sequence', 
                                                  'globalMetricValue': 'global_metric_value',
                                                  'Domain_architecture': 'domain_architecture', 
                                                  'InterPro': 'interpro', 
                                                  'Max_Topology': 'max_knot_topology', 
                                                  'Length': 'seq_length',
                                                  'FamilyName': 'family'})
df_test_reduced
```

## Save the cleaned data to a CSV file:

```python
df_test_reduced.to_csv(OUTPUT, index=False)
```
